```python
# Import libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Import Data
df = pd.read_csv(filepath_or_buffer="../data/crop_yield_train.csv")
```

```python
selected_cols = ["soil_moisture", "total_rainfall", "fertilizer_amount", "sunlight_hours", "field_id", "yield_tpha"]
```

```python
df_selected = df[selected_cols]
```

```python
df_selected
```

```python
from dataprocess import encodecategorical
```

```python
map_dict, data_update = encodecategorical(df_selected)
```

```python
data_update
```

```python
x = data_update.iloc[:, :-1]
y = data_update.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.4, random_state = 30)
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
```

```python
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(ytest, predictions)
```
