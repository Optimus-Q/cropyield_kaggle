```python
# Import libraries

import pandas as pd
import numpy as np
from dataprocess import encodecategorical
from xgboost import XGBRegressor
from matplotlib import pyplot as plt
```

```python
# Import Data
df = pd.read_csv(filepath_or_buffer="../data/crop_yield_test.csv")
```

```python
df
```

```python
df_copy = df.copy()
```

```python
df_copy
```

```python
selected_cols = ["soil_moisture", "total_rainfall", "fertilizer_amount", "sunlight_hours", "field_id"]
```

```python
df_selected = df_copy[selected_cols]
```

```python
df_selected
```

```python
map_dict, data_update = encodecategorical(df_selected)
```

```python
data_update
```

```python
model = XGBRegressor()
model.load_model("xgb_base_model.json")
```

```python
model
```

```python
predictions = model.predict(data_update)
```

```python
predictions
```

```python
pred_df = pd.DataFrame()
pred_df["id"] = df["id"].to_list()
pred_df["yield_tpha"] = predictions.tolist()
```

```python
pred_df
```

```python
pred_df.to_csv("baseline_pred.csv", index=False)
```
